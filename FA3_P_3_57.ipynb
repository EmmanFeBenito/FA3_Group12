```python
import time
import random

def time_sorted(data):
    start_time = time.time()
    sorted_data = sorted(data)
    end_time = time.time()
    return end_time - start_time

def generate_test_data(size):
    """Generate a random list of integers of the given size."""
    return [random.randint(1, 10000) for _ in range(size)]

sizes = [100, 500, 1000, 2000, 5000, 10000, 20000]
results = {}

for size in sizes:
    data = generate_test_data(size)
    elapsed_time = time_sorted(data)
    results[size] = elapsed_time

for size, elapsed_time in results.items():
    print(f"Size: {size}, Time: {elapsed_time:.6f} seconds")
```

```python
import matplotlib.pyplot as plt
import numpy as np

sizes = list(results.keys())
times = list(results.values())
log_times = [size * np.log2(size) for size in sizes]

plt.figure(figsize=(10, 5))
plt.plot(sizes, times, label='Measured Time', marker='o')
plt.plot(sizes, log_times, label='Theoretical O(n log n)', linestyle='--')
plt.xlabel('Input Size (n)')
plt.ylabel('Time (seconds)')
plt.title('Performance of Pythonâ€™s sorted() vs O(n log n)')
plt.legend()
plt.grid()
plt.show()
```
